# To Do that above conversion we can create helper function

```python
def convert(obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L
```

```python
# calling convert function by passing ---> 
convert('[{"movie_id": 28, "name": "Action"}, {"movie_id": 12, "name": "Adventure"}, {"movie_id": 14, "name": "Fantasy"}, {"movie_id": 878, "name": "Science Fiction"}]')
```

# the genre we have provided in convert function contains string thats why it is not working 

# now the genres are string so in order to use that we have to convert that strings to integer by using ast module ans that ast module containsa function literal_eval which does that work


```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'  )  # passing the same genre here
```

```python
def convert(obj): # here object will be string which we don't want
    L = []
    for i in ast.literal_eval(obj):   # by writing this --> ast.literal_eval(obj)   the string will be converted to list or integer
        L.append(i['name'])           # and then we can iterate loop on that list
    return L    
```

```python
#now we are sorting the cast so we have to pick only top 3 actror and rest we have to delete 
#by the same function that we use on top both genres or keyword but adding some modification using counter we use that to shown top 3 actor on movies
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:  # if else and counter because we want only 3 actor
            L.append(i['name']) # if will run three times for counter=0,1,2 and for 3 it will go to else and breakes
            counter+=1
        else:
            break
    return L
```

# coming from below to above we will apply steming to the data of tags column ,,,, steming means it converts different similar words into same words like loved,loving,love --> into --> love,love,love

```python
# for steming we need to install nltk library
!pip install nltk
```

```python
# for steming we need to import nltk library
import nltk
```

```python
from nltk.stem.porter import PorterStemmer # importing PorterStemmer class from nltk.stem.porter library
ps = PorterStemmer()  # creating object ps of porterstemmer
```

```python
# stem or steming will give only one word for different spelling of words with same meaning 
ps.stem('love')
ps.stem('loving')
ps.stem('loved')
```

```python
# stem or steming will give only one word for different spelling of words with same meaning 
ps.stem('dance')
ps.stem('dancing') 
ps.stem('danced')
```

```python
# Creating helper function of name stem where text will be input 
def stem(text):
    y = []       # creating a list 
    for i in text.split(): # converting string to list using text.split()
        y.append(ps.stem(i)) # steming the text using ps.stem(i) and storing it in y using y.append
    return " ".join(y)  # converting the list to string using ->" ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem) # applying the stem function to the tags column data in new_df dataset
```

## now we are creating a function named recommend so if someones inputs a movie then it will recommend the 5 similar movies from index


```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] # this line will find index value from of movies(vector) from table
    distances = similarity[movie_index] # measuring distance of array
    # For next step we sort the vectors values of each movie vector with emurate function so the index of movie's will not be loosed 
    # reverse=True is used for reverse the list or vectors values for similar recommendation and then its will give 5 simmilar cordinate suggestion using lambda function
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]  # key=lambda x:x[1])[1:6] --> this is used for recommending the similar 5 movies

    # below 2 lines are used for printing the index of 5 similar movies
    for i in movies_list:
        print(i[0])
```

```python
recommend('Avatar')
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] # this line will find index value from of movies(vector) from table
    distances = similarity[movie_index] # measuring distance of array
    # For next step we sort the vectors values of each movie vector with emurate function so the index of movie's will not be loosed 
    # reverse=True is used for reverse the list or vectors values for similar recommendation and then its will give 5 simmilar cordinate suggestion using lambda function
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]  # key=lambda x:x[1])[1:6] --> this is used for recommending the similar 5 movies

    # below 2 lines will print the names of 5 similar movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
